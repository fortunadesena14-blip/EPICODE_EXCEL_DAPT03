```python
# esercizio 8
# Scrivere un programma che • 
# in input acquisisce una lista di numeri e un numero K 
# in output, dovrà restituire la media di tutti i numeri nella lista maggiori o uguali a K 
# se non ce ne dovesse essere nessuno, dovrà stampare a schermo un messaggio adeguato.
```

```python
import numpy as np

lista = np.array([14, 28, 100, 46, 99, 67, 1, 9, 456, 90, 55, 43, 80, 7])
K = float(input("Inserisci a piacere un numero K: "))

filtro = lista[lista >= K]

if len(filtro) == 0:
    print("Dato il numero K scelto, nessun numero della lista è maggiore o uguale a K, per cui non è possibile calcolare la media")
else:
    media_K = filtro.mean()
    print("La media dei valori maggiori o uguali a", K, "è", media_K)
```

```python
# esercizio 13
# vogliamo dividere gli studenti in due squadre per un campionato di Uno nel seguente modo: 
# selezioneremo i nomi in posizione pari per un squadra, e i nomi in posizione dispari per l'altra.
import numpy as np

studenti = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry", "Isabelle", "John"]
nome = np.array(studenti)

squadra_pari = []
squadra_dispari = []

for i in range(len(nome)):
    if i % 2 == 0:
        squadra_pari.append(nome[i])
    else:
        squadra_dispari.append(nome[i])

print("Squadra pari:", squadra_pari)
print("Squadra dispari:", squadra_dispari)
```

```python
# esercizio 14
# dobbiamo confrontare, stampando tutto a video, il guadagno di ogni mese con la media dei guadagni precedenti, 
# e specificare nell'output se il guadagno attuale è maggiore o minore della media dei precedenti.
guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

```

```python
for i in range(len(guadagni)):
    if i == 0:
        print(f"Mese {i+1}: guadagno = {guadagni[i]} (nessun mese precedente)")
    else:
        media_precedenti = sum(guadagni[:i]) / i
        if guadagni[i] > media_precedenti:
            confronto = "maggiore"
        elif guadagni[i] < media_precedenti:
            confronto = "minore"
        else:
            confronto = "uguale"
        print(f"Mese {i+1}: guadagno = {guadagni[i]}, media precedenti = {media_precedenti:.2f} → {confronto}")
```
