```python
# media di charges rispetto a smoker
media2 = df.groupby("smoker")["charges"].mean()
print(round(media2,2))

# media di charges rispetto al genere
media3 = df.groupby("sex")["charges"].mean()
print(round(media3,2))
```

```python
round(df["bmi"].describe(),3)
```

```python
# Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi? 
Q1 = df["bmi"].quantile([0.25])
Q2 = df["bmi"].quantile([0.5])
Q3 = df["bmi"].quantile([0.75])
print(Q1)
print(Q2)
print(Q3)
quantili = df["bmi"].quantile([0.25, 0.50, 0.75])
print(quantili)
```

```python
df["quartile_bmi"] = pd.qcut(df["bmi"], 4, labels=["Q1","Q2","Q3","Q4"])
# bisogna aggiungere la colonna dei quantili al dataframe 
# in questo modo ho: da 0 a 0.24;0.25 a 0.49; 0.50 a 0.74; 0.75 a 1
misure = df.groupby("quartile_bmi", observed=False)["charges"].agg(["min","mean","max"])
print(round(misure,2)) # prendo solo 2 cifre dopo la virgola
```

```python
# esercizio dataframe
employees_df = pd.DataFrame({ 
    'employee_id': [101, 102, 103, 104, 105], 
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'], 
    'department_id': [1, 2, 1, 2, 3] })
```

```python
departments_df = pd.DataFrame({ 
    'department_id': [1, 2, 3], 
    'department_name': ['HR', 'IT', 'Finance'], 
    'location': ['New York', 'San Francisco', 'Chicago'] })
```

```python
# 1. Unire questi DataFrame in base alla colonna comune department_id,
# in modo da avere nel risultato informazioni sia sui dipendenti che sui dipartimenti, usando la funzione .merge()
df_unito = pd.merge(employees_df, departments_df, on = "department_id", how = "inner")
print(df_unito)
```

```python
# Per ogni DataFrame, trasformare la colonna department_id nell'indice, 
# facendo in modo che la modifica sia permanente; poi unire i due dataset mediante il metodo .join()
employees_df.set_index("department_id").join(
    departments_df.set_index("department_id"),
    how="inner"
)
```

```python
'''
merge() lavora sulle colonne, quindi l’indice originale del DataFrame non cambia.
Quando fai il print, vedi gli injoin() lavora sugli indici, quindi se prima hai fatto set_index("department_id"), il department_id diventa l’indice del nuovo DataFrame.

Gli indici non sono più 0, 1, 2…, ma i valori della colonna department_id.
Le altre colonne (employee_id, name, ecc.) rimangono come colonne normali.dici 0, 1, 2… come in un normale DataFrame, e le colonne employee_id, name, department_id ecc.
'''
```
