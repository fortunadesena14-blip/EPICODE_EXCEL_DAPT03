```python
# esercizio 3
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv("credenziali.env", override=True)

username = os.getenv("STUDENTE_DAPT")
password = os.getenv("PASSWORD")
host = os.getenv("HOST")
dbname = os.getenv("DBNAME")

print("USERNAME:", username)
print("PASSWORD:", password)
print("HOST:", host)
print("DBNAME:", dbname)


connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
print("Connection string:", connection_string)


db_engine = sqlalchemy.create_engine(connection_string)
```

```python
query2 = "SELECT * FROM dimcustomer"
df2 = pd.read_sql(query2,db_engine)
df2
```

```python
# Trasformiamo i nomi dei clienti in modo che abbiano solo lettere minuscole, e i cognomi in modo che abbiano solo lettere maiuscole
df2["FirstName"].str.lower()
```

```python
df2["LastName"].str.upper()
```

```python
# Sulla colonna EmailAddress, utilizzando il metodo .str.split(), estraiamo nome utente e dominio
df2.head()
```

```python
df2.columns
```

```python
print(df2.columns.tolist())
```

```python
df2.columns = df2.columns.str.strip()  # rimuove spazi iniziali/finali
```

```python
# Split su '@' per ottenere Username e Dominio
df2[['Username', 'Dominio']] = df2['EmailAddress'].str.split('@', expand=True)

# Controllo
print(df2[['EmailAddress', 'Username', 'Dominio']].head())
```

```python
df2["Phone"]
```

```python
# Split dei numeri di telefono sugli spazi
df2['Phone_parts'] = df2['Phone'].str.split(' ')

print(df2[['Phone', 'Phone_parts']].head())
```

```python
# Filtra le righe che contengono "21"
emails_21 = df2[df2['EmailAddress'].str.contains('21')]

# Mostra il risultato
print(emails_21[['EmailAddress']])
```

```python
# Estraiamo tutti gli indirizzi e-mail che contengono il numero "20" oppure il numero "10" 
emails_20_10 = df2[df2['EmailAddress'].str.contains('20|10', na=False)]
print(emails_20_10[['EmailAddress']])
```

```python
# Calcolare la lunghezza di ogni indirizzo e-mail ed estrarre i cinque più lunghi e i cinque più corti 
# Calcolo lunghezza
df2['EmailLength'] = df2['EmailAddress'].str.len()

# Cinque più lunghe
top5_longest = df2.nlargest(5, 'EmailLength')
print(top5_longest[['EmailAddress', 'EmailLength']])

# Cinque più corte
top5_shortest = df2.nsmallest(5, 'EmailLength')
print(top5_shortest[['EmailAddress', 'EmailLength']])
```

```python
# Modificare il dominio degli indirizzi e-mail da "adventure-works.com" a "aw-db.com" mediante il metodo .str.replace() 
df2['EmailAddress'] = df2['EmailAddress'].str.replace('adventure-works.com', 'aw-db.com', regex=False)
```

```python
#  Dalla colonna AddressLine1 estraiamo tutti gli indirizzi che contengono la sottostringa "Street"
street_addresses = df2[df2['AddressLine1'].str.contains('Street', na=False)]
print(street_addresses[['AddressLine1']])
```

```python
# Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp
import pandas as pd

# Conversione della colonna in formato Timestamp
fb['status_published'] = pd.to_datetime(fb['status_published'])
```

```python
# Estrarre informazioni temporali
fb['status_published'].dt.year       # anno
```

```python
fb['status_published'].dt.month      # mese (1-12)
```

```python
fb['status_published'].dt.dayofweek  # giorno della settimana (0=Lunedì, 6=Domenica)
```

```python
fb['status_published'].dt.dayofyear  # giorno dell'anno (1-365/366)
```

```python
# estrazione post del 2012
posts_2012 = fb[fb['status_published'].dt.year == 2012]
print(posts_2012)
```

```python
# estrazione post 2018
posts_maggio2018 = fb[
    (fb['status_published'].dt.year == 2018) &
    (fb['status_published'].dt.month == 5)
]
print(posts_maggio2018)
```

```python
#⃣  Confrontiamo il numero di post nei weekend vs resto della settimana
weekend_posts = fb[fb['status_published'].dt.dayofweek >= 5]  # sabato e domenica
weekday_posts = fb[fb['status_published'].dt.dayofweek < 5]

print("Post nei weekend:", len(weekend_posts))
print("Post nei giorni feriali:", len(weekday_posts))
```

```python
# Raggruppa per anno e calcola min e max della data
first_last_post = fb.groupby(fb['status_published'].dt.year)['status_published'].agg(['min', 'max'])
print(first_last_post)
```

```python
import os
import pandas as pd

# Cartella dei dataset
folder_path = r"C:\Users\fortu\OneDrive\Documenti\file\beginner_datasets"

# Lista per memorizzare i file con valori nulli
files_with_nulls = []

# Ciclo su tutti i file nella cartella
for file_name in os.listdir(folder_path):
    
    # Consideriamo solo i file .csv
    if file_name.endswith(".csv"):
        file_path = os.path.join(folder_path, file_name)
        
        # Leggi il CSV in un DataFrame
        df = pd.read_csv(file_path)
        
        # Conta i valori nulli per colonna e poi somma tutto
        total_nulls = df.isna().sum().sum()
        
        # Se ci sono valori nulli, aggiungiamo il nome del file alla lista
        if total_nulls > 0:
            files_with_nulls.append(file_name)

# Stampiamo i file che contengono valori nulli
print("File CSV con valori nulli:", files_with_nulls)
```
