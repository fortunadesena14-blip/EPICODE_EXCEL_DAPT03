```python
# esercio 3
# insurance
import pandas as pd
df = pd.read_csv(r"C:\Users\fortu\OneDrive\Documenti\file\insurance\insurance.csv")
df.head() # prime 5 righe
```

```python
df.info()
```

```python
print(df.columns)
```

```python
df.describe()
```

```python
# descrittori generali di BMI
print(df["bmi"].describe())
```

```python
# per non usaee groupby

print(df["region"].unique())
df.loc[df['region']=='southeast', 'charges'].mean()
df.loc[df['region']=='southwest', 'charges'].mean()
df.loc[df['region']=='northeast', 'charges'].mean()
df.loc[df['region']=='northwest', 'charges'].mean()


print(df["smoker"].unique())
df.loc[df['smoker']=='yes', 'charges'].mean()
df.loc[df['smoker']=='no', 'charges'].mean()

print(df["sex"].unique())
df.loc[df['sex']=='female', 'charges'].mean()
df.loc[df['sex']=='male', 'charges'].mean()


# quartili
quartili = pd.qcut(df['bmi'], 4)   # divide in 4 quartili
for q in quartili.unique():
    subset = df[quartili == q]
    print(f"{q}: min={subset['charges'].min()}, mean={subset['charges'].mean()}, max={subset['charges'].max()}")


# oppure
for reg in df['region'].unique():
    subset = df[df['region'] == reg]      # filtro per regione
    media = subset['charges'].mean()      # calcolo media
    print(f"{reg}: {media}")
    
for s in df['smoker'].unique():
    subset = df[df['smoker'] == s]
    print(f"{s}: {subset['charges'].mean()}")

for sex in df['sex'].unique():
    subset = df[df['sex'] == sex]
    print(f"{sex}: {subset['charges'].mean()}")
```

```python
conteggio = df["Age (years)"].value_counts()
print(conteggio)
```

```python
#Selezioniamo i dati relativi a diverse fasce di età: 20-29, 30-39, 40-50;
fascia_20_29 = df[(df["Age (years)"] >= 20) & (df["Age (years)"]<= 29)]
fascia_30_39 = df[(df["Age (years)"] >= 30) & (df["Age (years)"]<= 39)]
fascia_40_50 = df[(df["Age (years)"] >= 40) & (df["Age (years)"]<= 50)]
```

```python
#Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?
print(fascia_20_29["Diastolic blood pressure (mm Hg)"].mean())
print(fascia_30_39["Diastolic blood pressure (mm Hg)"].mean())
print(fascia_40_50["Diastolic blood pressure (mm Hg)"].mean())
```

```python
# nomi unici dei leggendari
# seleziono solo i leggendari e prendo i nomi unici
nomi_leggendari = df.loc[df["Legendary"], "Name"].unique()
print(nomi_leggendari)

```

```python
# legendary e tipe 1 = grass
leggendari = df["Legendary"]  # già True/False
grass = df["Type 1"] == "Grass"
nomi = df.loc[leggendari & grass, "Name"].unique()
print(nomi)
```

```python
leggendari = df["Legendary"]  # già True/False
ice = df["Type 1"] == "Ice"
fire = df["Type 1"] == "Fire"
nomi = df.loc[leggendari & grass | fire, "Name"].unique()
print(nomi)
```

```python
# • Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
prima_gen = df["Generation"] == 1
attacco = df["Attack"] > 50
hp = df["HP"] < 60
filtro = prima_gen & attacco & hp
pokemon = df.loc[filtro, "Name"].unique()
print(pokemon)
```
